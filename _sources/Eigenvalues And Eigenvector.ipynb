# Eigenvalues And Eigenvector
### Definisi
* Vektor eigen (eigenvector) atau vektor karakteristik dari suatu matriks berukuran n × n adalah vektor tak nol yang hanya mengalami perubahan panjang ketika dikali dengan matriks tersebut.
* Nilai eigen (eigenvalue) yang berasosiasi dengan vektor tersebut, umumnya dilambangkan dengan λ, menyatakan besar perubahan panjang vektor yang terjadi.

### Mencari Eigenvalues And Eigenvector
##### Eigenvalues 
* Susun persamaan karakteistik: det(A - λI) = 0, di mana A adalah matriks yang diberikan, λ (lambda) adalah nilai eigen yang akan dicari, dan I adalah matriks identitas dengan ukuran yang sama dengan A.
* Selesaikan persamaan karakteristik tersebut untuk mencari nilai λ. Ini akan menghasilkan persamaan polinomial dalam λ.
* Cari akar-akar polinomial tersebut. Akar-akar tersebut adalah nilai eigen dari matriks A.

#### Eigenvector
* Untuk setiap nilai eigen yang telah ditemukan, substitusikan nilai λ ke dalam persamaan (A - λI)x = 0, di mana x adalah vektor eigen yang akan dicari.
* Selesaikan sistem persamaan linier tersebut untuk mencari vektor x yang memenuhi persamaan. Solusi yang tidak trivial (tidak semuanya nol) adalah vektor eigen yang bersesuaian dengan nilai eigen tersebut.
* Anda dapat memilih satu vektor eigen untuk setiap nilai eigen, dan vektor eigen lainnya dapat diperoleh dengan mengalikan vektor eigen tersebut dengan skalar (bilangan).

## Mencari Eigen Values



Jawab:

Diberikan matriks:

$$A = \begin{bmatrix}
3 & 0 \\
0 & 3
\end{bmatrix}
, \quad
B = \begin{bmatrix}
-2 & 0 \\
0 & 4
\end{bmatrix}$$


Eigenvalues Matriks A

Kita cari $\lambda$ yang memenuhi:


$$\det(A - \lambda I) = 0$$

Dengan matriks identitas

$$I = \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}$$


Maka,

$$A - \lambda I = \begin{bmatrix}
3 - \lambda & 0 \\
0 & 3 - \lambda
\end{bmatrix}$$

Determinannya:

$$\det(A - \lambda I) = (3 - \lambda)(3 - \lambda) = (3 - \lambda)^2$$

Persamaan karakteristik:

$$(3 - \lambda)^2 = 0 \implies \lambda = 3$$

Jadi, nilai eigen dari A adalah:

$$\boxed{ \lambda_1 = 3, \quad \lambda_2 = 3 }$$

Eigenvalues Matriks B

Cari $\lambda$ yang memenuhi:

$$\det(B - \lambda I) = 0$$

$$B - \lambda I = \begin{bmatrix}
-2 - \lambda & 0 \\
0 & 4 - \lambda
\end{bmatrix}$$

Determinannya:

$$\det(B - \lambda I) = (-2 - \lambda)(4 - \lambda)$$

Persamaan karakteristik:
$$(-2 - \lambda)(4 - \lambda) = 0$$

Sehingga,

$$\lambda_1 = -2, \quad \lambda_2 = 4$$

Urutan Eigenvalues dari yang terbesar ke terkecil

Gabungkan semua nilai eigen:

$$\{3, 3, -2, 4\}$$

Diurutkan menjadi:
$$\boxed{
4, \quad 3, \quad 3, \quad -2
}$$


```python
import numpy as np

# Definisikan matriks A dan B
A = np.array([[3, 0],
              [0, 3]])

B = np.array([[-2, 0],
              [0, 4]])

# Hitung nilai eigen untuk matriks A
eigenvalues_A, _ = np.linalg.eig(A)

# Hitung nilai eigen untuk matriks B
eigenvalues_B, _ = np.linalg.eig(B)

# Gabungkan semua nilai eigen
all_eigenvalues = np.concatenate((eigenvalues_A, eigenvalues_B))

# Urutkan nilai eigen dari yang terbesar ke yang terkecil
sorted_eigenvalues = np.sort(all_eigenvalues)[::-1]

# Tampilkan hasil
print("Nilai eigen dari matriks A:", eigenvalues_A)
print("Nilai eigen dari matriks B:", eigenvalues_B)
print("Urutan nilai eigen (dari yang terbesar ke terkecil):", sorted_eigenvalues)
```

## Mencari Eigen Vektor

jawab:

### Vektor Eigen Matriks A dan B

### Vektor Eigen Matriks A

Dari perhitungan sebelumnya, kita mendapatkan nilai eigen dari matriks A:

$$
\lambda = 3 \quad \text{(dengan multiplicity 2)}
$$

Untuk mencari vektor eigen, kita perlu menyelesaikan persamaan:

$$
(A - \lambda I) \mathbf{v} = 0
$$

Dengan:

$$
A - 3I =
\begin{bmatrix}
3 - 3 & 0 \\
0 & 3 - 3
\end{bmatrix}
=
\begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}
$$

Maka:

$$
\begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

Setiap vektor dalam bentuk:

$$
\mathbf{v} =
\begin{bmatrix}
1 \\
0
\end{bmatrix}, \quad
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

adalah vektor eigen yang sesuai dengan nilai eigen:

$$
\lambda = 3
$$

Jadi, vektor eigen untuk matriks A adalah:

$$
\boxed{
\begin{bmatrix}
1 \\
0
\end{bmatrix}, \quad
\begin{bmatrix}
0 \\
1
\end{bmatrix}
}
$$

### Vektor Eigen Matriks B

Dari perhitungan sebelumnya, kita mendapatkan nilai eigen dari matriks B:

$$
\lambda_1 = -2, \quad \lambda_2 = 4
$$

### Untuk nilai eigen $\lambda_1 = -2$:

$$
B + 2I =
\begin{bmatrix}
-2 + 2 & 0 \\
0 & 4 + 2
\end{bmatrix}
=
\begin{bmatrix}
0 & 0 \\
0 & 6
\end{bmatrix}
$$

$$
\begin{bmatrix}
0 & 0 \\
0 & 6
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

Maka:

$$
\mathbf{v} =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

adalah vektor eigen untuk:

$$
\lambda_1 = -2
$$

### Untuk nilai eigen $\lambda_2 = 4$:

$$
B - 4I =
\begin{bmatrix}
-2 - 4 & 0 \\
0 & 4 - 4
\end{bmatrix}
=
\begin{bmatrix}
-6 & 0 \\
0 & 0
\end{bmatrix}
$$

$$
\begin{bmatrix}
-6 & 0 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

Maka:

$$
\mathbf{v} =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

adalah vektor eigen untuk:

$$
\lambda_2 = 4
$$
