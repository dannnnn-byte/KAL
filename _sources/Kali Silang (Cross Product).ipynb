# Perkalian Silang (Cross Product)
**Definisi:**
Perkalian silang (cross product) adalah operasi antara dua vektor di ruang tiga dimensi ($\mathbb{R}^{3}$) yang menghasilkan vektor baru yang:
* Tegak lurus terhadap kedua vektor asal,
* Arah ditentukan oleh aturan tangan kanan (right-hand rule),
* Panjang vektor hasilnya sebesar:

    $|\vec{u} \times \vec{v}| = |\vec{u}| \cdot |\vec{v}| \cdot \sin \theta$

    di mana $\theta$ adalah sudut antara $\vec{u}$ dan $\vec{v}$.

## Definisi Hasil Kali Silang
Misalkan $\vec{u} =
\begin{bmatrix}
\mathbf{u_1} \\
\mathbf{u_2} \\
\mathbf{u_3}
\end{bmatrix}
.$ dan $\vec{v} =
\begin{bmatrix}
\mathbf{v_1} \\
\mathbf{v_2} \\
\mathbf{v_3}
\end{bmatrix}$ adalah vektor-vektor di $\mathbb{R}^{3}$. 
**Hasil kali silang** dari u dan v dinotasikan sebagai $\vec{u} \times \vec{v}$, adalah vektor.

$\vec{u} \times \vec{v} =
\begin{bmatrix}
\mathbf{u_2 v_3 - u_3v_2} \\
\mathbf{-(u_1v_3 -u_3v_1)} \\
\mathbf{u_1 v_2 - u_2v_1}
\end{bmatrix}$

## Contoh Menghitung Hasil Kali Silang
Misalkan $\vec{u} =
\begin{bmatrix}
2 \\
-1 \\
4
\end{bmatrix}$ dan $\vec{v} =
\begin{bmatrix}
3 \\
2 \\
5
\end{bmatrix}$ temukan $\vec{u} \times \vec{v}$, dan verifikasi bahwa hasil kali silang ini tegak lurus terhadap baik $\vec{u}$ maupun $\vec{v}$.

Menggunakan **Definisi**

$\vec{u} \times \vec{v} =
\begin{bmatrix}
\mathbf{u_2 v_3 - u_3v_2} \\
\mathbf{-(u_1v_3-u_3v_1)} \\
\mathbf{u_1 v_2 - u_2v_1}
\end{bmatrix} =
\begin{bmatrix}
(-1)(5) - (4)(2) \\
-((2)(5) - (4)(3))\\
(2)(2) - (-1)(3)
\end{bmatrix} =
\begin{bmatrix}
-5 -8 \\
-(10-12)\\
4+3
\end{bmatrix}=
\begin{bmatrix}
-13 \\
2\\
7
\end{bmatrix}$

**Jadi:**

$\vec{u} \times \vec{v} = \begin{bmatrix}
-13 \\
2\\
7
\end{bmatrix}$

**Verifikasi: Tegak Lurus**
Dua vektor tegak lurus jika hasil dot product-nya nol:
1. Cek apakah $\vec{u} \cdot (\vec{u} \times \vec{v}) = 0$:

    $\vec{u} \cdot (\vec{u} \times \vec{v}) =\begin{bmatrix}
2 \\
-1 \\
4
\end{bmatrix} \cdot
\begin{bmatrix}
-13 \\
2\\
7
\end{bmatrix}
=(2)(−13)+(−1)(2)+(4)(7)=−26−2+28=0$

    Tegak lurus terhadap $\vec{u}$.

2. Cek apakah $\vec{v} \cdot (\vec{u} \times \vec{v}) = 0$:

    $\vec{v} \cdot (\vec{u} \times \vec{v}) =\begin{bmatrix}
3 \\
2 \\
5
\end{bmatrix} \cdot
\begin{bmatrix}
-13 \\
2\\
7
\end{bmatrix}
=(3)(−13)+(2)(2)+(5)(7)=−39+4+35=0$
 
     Tegak lurus terhadap $\vec{v}$.
 
**Kesimpulan:**
Hasil kali silang dari

$\vec{u} =
\begin{bmatrix}
2 \\
-1 \\
4
\end{bmatrix}$ dan $\vec{v} =
\begin{bmatrix}
3 \\
2 \\
5
\end{bmatrix}$

adalah:

$\vec{u} \times \vec{v} = \begin{bmatrix}
-13 \\
2\\
7
\end{bmatrix}$

dan hasil ini tegak lurus terhadap kedua vektor.
